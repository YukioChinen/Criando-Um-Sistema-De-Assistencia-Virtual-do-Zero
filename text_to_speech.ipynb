```python
from gtts import gTTS

text_to_say = "Hello World!"

language = "en"

gtts_object = gTTS(text = text_to_say, lang = language, slow = False)

gtts_object.save("hello_world.mp3")
```

```python
from IPython.display import Audio

Audio("/content/hello_world.mp3")
```

```python
texto = "Aqui é Brasil!"

lingua = "pt"

gtts_object = gTTS(text = texto, lang = lingua, slow = False)

gtts_object.save("brasil.mp3")
```
